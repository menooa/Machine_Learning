```python
import pandas as pd

df = pd.read_csv('data-lab-6-ckd-courseproject.csv')
display(df)
```

```python
print(df.isnull().sum())
print('')
display(df[df.duplicated(keep=False)])
```

```python
# create GFR and CKD Stages columns
def calc_gfr(Scr, Age):
    return (175 * (Scr) ** -1.154) * (Age ** -0.203) # GFR (mL/min/1.73 m^2) = 175 × (Scr)^-1.154 × (Age)-0.203

def ckd_stages(gfr):
    bins = [0, 45, 90, 250]
    labels = [3, 2, 1]  # number of items should be 1 less than bins
    return pd.cut(gfr, bins=bins, labels=labels)
    
df['GFR'] = calc_gfr(df['Serum Creatinine'], df['Age'])
df['CKD Stages'] = ckd_stages(df['GFR'])

df['CKD Stages'].value_counts()
display(df.head())
```

```python
for column in df.select_dtypes(include=['number']).columns:  # can also use df.select_dtypes(include=['number'])
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_band = Q1 - IQR
    upper_band = Q3 + IQR

    outlier_list = [i for i in df[column] if i < lower_band or i > upper_band]
    print(f'{column}: {outlier_list}')
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder

x = df.drop(['Class', 'CKD Stages'], axis=1)
y = df['CKD Stages']

x_num = x.select_dtypes(include=['number'])  # includes both integers and floats
x_cat = x.select_dtypes(include=['object'])

transformer = ColumnTransformer([
    ('num', StandardScaler(), x_num.columns),
    ('cat', OneHotEncoder(), x_cat.columns)  # we can also use OrdinalEncoder, but not LabelEncoder, it does work in ColumnTransformer, casue it is made for 1D array for target
])

x = transformer.fit_transform(x)
```

```python
# create back a table to get correlation

x_names = transformer.get_feature_names_out()  # Get feature names from the transformer
transformed_df = pd.DataFrame(x, columns=x_names)
transformed_df['CKD Stages'] = y.values
transformed_df.head()
```

```python
# corr table
corr_x_y = transformed_df.corr()['CKD Stages'].sort_values(ascending=False)
corr_x_y
```

```python
# get rows with absolute value > 0.3
core_above3 = corr_x_y[abs(corr_x_y) > 0.3]
core_above3
```

```python
final_x_names = core_above3.index.tolist()
final_x_names
```

```python
final_x = transformed_df[final_x_names].drop(['CKD Stages'], axis=1)
final_y = transformed_df['CKD Stages']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(final_x, final_y, stratify=final_y, test_size=0.2, random_state=42)
# stratify=final_y: Ensures the class distribution (e.g., ratios of CKD stages) is preserved in both train and test sets. Critical for imbalanced datasets.
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
y_pred = log_reg.predict(x_test)
acc_score = accuracy_score(y_test, y_pred)
acc_score
```

```python
from sklearn.ensemble import RandomForestClassifier

rand_for = RandomForestClassifier()
rand_for.fit(x_train, y_train)
y_pred = rand_for.predict(x_test)
acc_score1 = accuracy_score(y_test, y_pred)
acc_score1
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
acc_score2 = accuracy_score(y_test, y_pred)
acc_score2
```
