```python
# LLM
# Model Type: Sequence-to-Sequence (Seq2Seq) Transformer
# Architecture: Encoder-Decoder Transformer
# Example Models Used: "facebook/blenderbot-400M-distill"
# Primary Function: Conversational AI (Chatbot)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
# transformers: A popular library by Hugging Face for working with pre-trained NLP models.
# AutoTokenizer: Automatically selects the correct tokenizer for a given model.
# AutoModelForSeq2SeqLM: Loads a sequence-to-sequence model (good for tasks like text generation, chatbots, summarization, translation, etc.).

model_name = "facebook/blenderbot-400M-distill"
# A modle available in hugging face: A lightweight version of Facebook's BlenderBot, a chatbot model optimized for conversations.
# other models in Hugging face: https://huggingface.co/models
# "google/flan-t5-small"
# "facebook/bart-base"

model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
# Downloads the pre-trained BlenderBot model from Hugging Face and Loads it into memory for inference.
tokenizer = AutoTokenizer.from_pretrained(model_name)
# Loads the Tokenizer
# A tokenizer converts raw text (words) into numerical tokens that the model understands):
# Example: tokenizer.encode("Hello, how are you?") => Output: [8794, 47, 172, 42, 312] (tokenized numbers)

```

```python
# define the chat function
def chat_with_bot():
    while True:
        input_text = input("You: ")

        if input_text.lower() in ["quit", "exit", "bye"]:
            print("Chatbot: Goodbye!")
            break

        inputs = tokenizer.encode(input_text, return_tensors="pt")  # encode the input text
        ouputs = model.generate(inputs, max_new_tokens=150)  # generate response (encoded)
        response = tokenizer.decode(ouputs[0], skip_tokens=True).strip() # decode

        print("Chatbot:", response)

chat_with_bot()  # start chatting
```
